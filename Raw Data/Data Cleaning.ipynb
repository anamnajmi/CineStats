```python
import pandas as pd
```

```python
# Load Dataset
df = pd.read_csv("imdb_top_1000.csv")
```

```python
# Head
df.head()
```

```python
rows, cols = df.shape
print(rows) #Entries
print(cols) #Categotries
```

```python
df.columns.tolist()
```

```python
# Picking only the columns that matter
df = df[['Series_Title', 'Genre', 'Director', 'Released_Year', 'Overview', 'IMDB_Rating','Poster_Link']]
df.head()
```

```python
# Renaming the selected columns for ease
df = df.rename(columns={
    'Series_Title':'title',
    'Genre':'genre',
    'Director':'director',
    'Released_Year':'year',
    'Overview':'description',
    'IMDB_Rating':'rating',
    'Poster_Link':'link',
})
```

```python
# Checking Nulls
df.isnull().sum()
```

```python
# Handling Missing Data
df = df.dropna(subset=['title', 'genre', 'description','link'])
```

```python
# Cleaning text cols
df['title'] = df['title'].str.strip()
df['genre'] = df['genre'].str.strip()
df['director'] = df['director'].str.strip()
df['description'] = df['description'].str.strip()
```

```python
# Adding Simple Reviews Column
df['reviews'] = df['title'].apply(lambda x: f"'{x}' is a must-watch movie loved by many!")
```

```python
# Save the file
df.to_csv("movies.csv", index=False)
print("Cleaned CSV Updated")
```
